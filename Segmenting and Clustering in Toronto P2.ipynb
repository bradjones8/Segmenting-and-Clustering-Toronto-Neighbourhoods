<p>Retrieve the Geographic Data and merge Latitude and Longitude columns with the Toronto Data Dataframe</p>

```python
import pandas as pd


torontoData = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
torontoData.head()
```

```python
torontoData = torontoData[~torontoData.Borough.str.contains("Not assigned")]
torontoData.rename(columns={"Postal Code": "PostalCode"}, inplace=True)

torontoData.head()
```

```python
torontoData.shape
```

```python
import io
import requests
url="http://cocl.us/Geospatial_data"
s=requests.get(url).content
geoData=pd.read_csv(io.StringIO(s.decode('utf-8')))

geoData.head()
```

```python
torontoData['Latitude'] = 0
torontoData['Longitude'] = 0
```

```python
torontoData.head()
```

<p>Matching the PostalCode of both dataframes and merging the Latitude and Longitude columns with torontoData below:</p> 

```python

for i in range(len(torontoData)):
    for j in range(len(geoData)):
        
        if (torontoData.iloc[i,0] == geoData.iloc[j,0]):
            torontoData.iloc[i,3] = round(geoData.iloc[j,1], 6)
            torontoData.iloc[i,4] = round(geoData.iloc[j,2], 6)
            break
torontoData.head(15)
```
